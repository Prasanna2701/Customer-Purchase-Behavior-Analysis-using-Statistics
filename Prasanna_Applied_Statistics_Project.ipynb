##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  

```python
# Lets first calculate measures of central tendency (mean, median, mode)
# 1 : Mean
key_variables = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
                 'MntFishProducts', 'NumDealsPurchases', 'MntGoldProds', 'NumWebPurchases', 'NumWebVisitsMonth']

means = df[key_variables].mean()

print('Means of selected variables in the dataset:')
for variable, mean_val in means.items():
    print(f"  {variable}: {mean_val:.2f}")
```

```python
# 2 : Median
key_variables = ['Income', 'Recency', 'Kidhome', 'Teenhome', 'MntWines', 'MntFruits',
                 'MntMeatProducts', 'MntFishProducts', 'NumDealsPurchases', 'NumWebPurchases', 'NumWebVisitsMonth']

medians = df[key_variables].median()

print('Medians of key variables in the dataset:')
for variable, median_val in medians.items():
    print(f"  {variable}: {median_val:.2f}")
```

```python
# 3 : Mode
key_variables = ['Income', 'Recency', 'Kidhome', 'Teenhome', 'MntWines', 'MntFruits',
                 'MntMeatProducts', 'MntFishProducts', 'NumDealsPurchases', 'NumWebPurchases', 'NumWebVisitsMonth']

print('Modes of key variables in the dataset:')
for variable in key_variables:
    mode_values = df[variable].mode()
    print(f"  {variable}:")
    for mode_value in mode_values:
        print(f"    - {mode_value}")
```

# Measures of dispersion (variance, standard deviation) for key variables:

```python
# 4 : Variance
key_variables = ['Income','Recency','Kidhome','Teenhome','MntWines','MntFruits','MntMeatProducts','MntFishProducts','NumDealsPurchases','NumWebPurchases','NumWebVisitsMonth']
variances = df[key_variables].var()

print('Variances of key variables in the dataset:')
for variable, var in variances.items():
    print(f"  {variable}: {var:.2f}")
```

```python
# 5 : Standard Deviation
key_variables = ['Income','Recency','Kidhome','Teenhome','MntWines','MntFruits','MntMeatProducts','MntFishProducts','NumDealsPurchases','NumWebPurchases','NumWebVisitsMonth']
standard_deviations = df[key_variables].std()

print('Standard deviations of key variables in the dataset:')
for variable, std_dev in standard_deviations.items():
    print(f"  {variable}: {std_dev:.2f}")
```

# Identify Outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create boxplots for key variables
key_variables = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
                 'NumDealsPurchases', 'MntGoldProds', 'NumWebPurchases', 'NumWebVisitsMonth']

# Set a consistent style for all plots
sns.set_theme(style="whitegrid")

# Define a color palette
palette = sns.color_palette("muted")

for i, variable in enumerate(key_variables):
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[variable], width=0.4, color=palette[i])
    plt.title(f'Distribution of {variable}', fontsize=14)
    plt.ylabel(variable, fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    sns.despine(left=True)

    median_val = df[variable].median()
    plt.axhline(y=median_val, color='red', linestyle='--', linewidth=1)
    plt.text(0.1, median_val + 5, f'Median: {median_val:.2f}', color='red')

    # Identify and annotate potential outliers (adjust threshold as needed)
    outlier_threshold = df[variable].quantile(0.75) + 1.5 * (df[variable].quantile(0.75) - df[variable].quantile(0.25))
    outliers = df[variable][df[variable] > outlier_threshold]
    if not outliers.empty:
        plt.scatter(x=[0]*len(outliers), y=outliers, color='red', marker='o', label='Potential Outliers')
        plt.legend()

    plt.show()
    print()
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

```python
total_purchases = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)
p_purchase = total_purchases.mean() / total_purchases.max()
expected_purchases = total_purchases.mean()
```

```python
from scipy.stats import norm

# Calculate income statistics
mu_income, sigma_income = df['Income'].mean(), df['Income'].std()
prob_income_gt_70000 = 1 - norm.cdf(70000, mu_income, sigma_income)
expected_income = mu_income + sigma_income * norm.ppf(prob_income_gt_70000)

# Calculate purchase probability and expected number (replace with your logic)
# Example: probability of purchase if income > 50000
p_purchase = df[df['Income'] > 50000]['TotalSpend'].mean() / df['TotalSpend'].mean()
expected_purchases = p_purchase * len(df)

print("Probability Distributions and Calculated Values:")
print(f"Total Purchases:")
print(f"   Probability of a purchase: {p_purchase:.4f}")
print(f"   Expected number of purchases: {expected_purchases:.2f}")

print(f"\nIncome:")
print(f"   Probability of income > $70,000: {prob_income_gt_70000:.4f}")
print(f"   Expected income: ${expected_income:.2f}")
```

## Task 4: Insights and Customer Segmentation üìà

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

**Deliverables**:

- **Key insights regarding relationships between variables and distinct customer segments based on behaviors.**

  

```python
df['TotalSpend'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Income', y='TotalSpend', data=df, color='skyblue', alpha=0.7)
plt.title('Relationship between Income and Total Spending', fontsize=14)
plt.xlabel('Income', fontsize=12)
plt.ylabel('Total Spending', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
sns.despine()

sns.regplot(x='Income', y='TotalSpend', data=df, scatter=False, color='red')

plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = df.select_dtypes(include=['number']).corr()

# Visualize correlation matrix using a heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5,
            annot_kws={"size": 10})
plt.title('Correlation Matrix', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
